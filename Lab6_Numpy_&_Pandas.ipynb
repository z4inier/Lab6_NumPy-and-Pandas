```python
# PROCEDURE 2: Array Operations
years_exp = np.array([1, 3, 5, 7, 10])  # reset for clean demo
 
# Original Code
exp_plus_5 = years_exp + 5
print("\nYears + 5:", exp_plus_5)
 
exp_times_2 = years_exp * 2
print("Years * 2:", exp_times_2)
 
sample1 = np.array([1, 2, 3])
sample2 = np.array([4, 5, 6])
dot_result = np.dot(sample1, sample2)
print("Dot Product:", dot_result)
```

```python
# ── Try It ────────────────────────────────────────────────
# Subtraction and division
exp_minus_1 = years_exp - 1
exp_div_2   = years_exp / 2
print("\n[TRY IT] Years - 1:", exp_minus_1)
print("[TRY IT] Years / 2:", exp_div_2)
 
# np.exp() and np.log()
print("[TRY IT] np.exp(years_exp):", np.exp(years_exp))
print("[TRY IT] np.log(years_exp):", np.log(years_exp))
 
```

```python
# PROCEDURE 3: Indexing and Slicing
years_exp = np.array([1, 3, 5, 7, 10])
salaries  = np.array([[50, 60, 70], [80, 90, 100]])
 
# Original Code
print("\nFirst year of experience:", years_exp[0])
print("First two salaries:", salaries[0, :2])
print("Second column salaries:", salaries[:, 1])
print("Last year of experience:", years_exp[-1])
```

```python
# PROCEDURE 4: Reshaping and Dimensions

# Original Code
reshaped_exp = np.reshape(np.arange(1, 7), (2, 3))
print("\nReshaped Experience Array:\n", reshaped_exp)
 
flattened_exp = reshaped_exp.flatten()
print("Flattened Array:", flattened_exp)
 
print("Transposed Array:\n", reshaped_exp.T)
```

```python
# PROCEDURE 5: Broadcasting in NumPy
salaries = np.array([[50, 60, 70], [80, 90, 100]])
 
# Original Code
bonus = np.array([5, 10, 15])
salaries_with_bonus = salaries + bonus
print("\nSalaries after bonus:\n", salaries_with_bonus)
```

```python
# ── Try It ────────────────────────────────────────────────
# Multiply by a scaling factor using broadcasting
scaling_factor = np.array([1.1, 1.2, 1.15])
scaled_salaries = salaries * scaling_factor
print("[TRY IT] Salaries after scaling:\n", scaled_salaries)
 
# Broadcast a scalar
salaries_plus_flat_bonus = salaries + 5
print("[TRY IT] All salaries + flat 5k:\n", salaries_plus_flat_bonus)
```

```python
# PROCEDURE 6: Statistical Operations

years_exp = np.array([1, 3, 5, 7, 10])
salaries  = np.array([[50, 60, 70], [80, 90, 100]])
 
# Original Code
print("\nMean experience:", np.mean(years_exp))
print("Std deviation of experience:", np.std(years_exp))
print("Max salary:", np.max(salaries), "| Min salary:", np.min(salaries))
print("Sum of all salaries:", np.sum(salaries))
```

```python
# ── Try It ────────────────────────────────────────────────
print("\n[TRY IT] Median of years_exp:", np.median(years_exp))
print("[TRY IT] 25th percentile:", np.percentile(years_exp, 25))
print("[TRY IT] 75th percentile:", np.percentile(years_exp, 75))
print("[TRY IT] Median of salaries:", np.median(salaries))
```

```python
# PROCEDURE 7: NumPy Functions on Arrays
 
# Original Code
angles = np.array([0, np.pi/4, np.pi/2])
print("\nSine of angles:", np.sin(angles))
print("Cosine of angles:", np.cos(angles))
 
salaries = np.array([[50, 60, 70], [80, 90, 100]])
salary_sums = np.apply_along_axis(np.sum, 1, salaries)
print("Sum of Salaries per person:", salary_sums)
```

```python
# ── Try It ────────────────────────────────────────────────
sample_arr = np.array([1, 4, 9, 16, 25])
print("\n[TRY IT] np.sqrt():", np.sqrt(sample_arr))
print("[TRY IT] np.log():", np.log(sample_arr))
 
# apply_along_axis with a custom function (column-wise max)
col_max = np.apply_along_axis(np.max, 0, salaries)
print("[TRY IT] Max per column:", col_max)
 
# Custom function: normalize a row (subtract min, divide by range)
def normalize(row):
    return (row - np.min(row)) / (np.max(row) - np.min(row))
 
normalized = np.apply_along_axis(normalize, 1, salaries)
print("[TRY IT] Row-normalized salaries:\n", normalized)
```

```python
# PROCEDURE 8: Integrating NumPy and Pandas
# ============================================================

# Original Code
np.random.seed(42)
data = np.random.randint(1, 50, size=(5, 3))
 
df_data = pd.DataFrame(data, columns=['X', 'Y', 'Z'])
print("\nGenerated Data:\n", df_data)
 
df_data['Log_X']  = np.log(df_data['X'])
df_data['Sqrt_Y'] = np.sqrt(df_data['Y'])
print("DataFrame with NumPy Transformations:\n", df_data)
 
print("Correlation Matrix:\n", df_data.corr(numeric_only=True))
```

```python
# ── Try It ────────────────────────────────────────────────
print("\n[TRY IT] Column Means via np.mean():")
print(np.mean(df_data[['X', 'Y', 'Z']].values, axis=0))
 
print("\n[TRY IT] Column Medians via np.median():")
print(np.median(df_data[['X', 'Y', 'Z']].values, axis=0))
 
df_data['Squared_Z'] = np.square(df_data['Z'])
df_data['Exp_X']     = np.exp(df_data['X'])
print("[TRY IT] DataFrame with np.square() and np.exp():\n",
      df_data[['X', 'Z', 'Squared_Z', 'Exp_X']])
 
```

```python
# PROCEDURE 9: Importing and Summarizing Data Using Pandas
 
# Original Code
df_data[['X', 'Y', 'Z']].to_csv('sample_data.csv', index=False)
print("\nData saved to 'sample_data.csv'.")
 
df_imported = pd.read_csv('sample_data.csv')
print("Imported DataFrame:\n", df_imported)
 
summary_stats = df_imported.describe()
print("Summary Statistics:\n", summary_stats)
 
print("Column Means:\n", df_imported.mean(numeric_only=True))
print("Column Standard Deviations:\n", df_imported.std(numeric_only=True))
 
```

```python
# ── Try It ────────────────────────────────────────────────
df_imported['Sum_XY'] = df_imported['X'] + df_imported['Y']
df_imported['Diff_ZX'] = df_imported['Z'] - df_imported['X']
print("\n[TRY IT] DataFrame with new calculated columns:\n", df_imported)
 
df_imported.to_csv('sample_data_modified.csv', index=False)
print("[TRY IT] Modified DataFrame saved to 'sample_data_modified.csv'.")
```

```python
# Step 1: Load the Stack Overflow 2023 Developer Survey dataset
df_kaggle = pd.read_csv('survey_results_public.csv')
print("Loaded Dataset:\n", df_kaggle.head())

# Step 2: Select relevant columns

df_subset = df_kaggle[['Country', 'EdLevel', 'YearsCodePro', 'ConvertedCompYearly']]
print("Subset of Data:\n", df_subset.head())

# Step 3: Clean the data by dropping rows with missing values
df_clean = df_subset.dropna()
print("Cleaned Data:\n", df_clean.head())

# Step 4: Convert YearsCodePro to numbers first
df_clean['YearsCodePro'] = pd.to_numeric(df_clean['YearsCodePro'], errors='coerce')

# Now you can categorize experience without the TypeError
df_clean['ExperienceLevel'] = np.where(df_clean['YearsCodePro'] >= 10, 'Senior', 'Junior')

# Step 5: Group by Country and ExperienceLevel
# (Note: Using 'ConvertedCompYearly' as we found earlier)
grouped_data = df_clean.groupby(['Country', 'ExperienceLevel'])['ConvertedCompYearly'].mean()
print("Grouped Average Salary:\n", grouped_data.head())

# Step 6: Reset index for readability
grouped_data = grouped_data.reset_index()
print("Formatted Grouped Data:\n", grouped_data.head())
```
